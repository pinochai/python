```python
####avocado percentage per type

type_perc = avocado.groupby(["TYPE"]).agg(VOLUME = ("TOTAL VOLUME" , "sum"))
print(type_perc)
colors = ['pink', 'steelblue']  
plot = type_perc.plot.pie(y= 'VOLUME', figsize=(10, 10)  , autopct='%1.0f%%' , colors=colors , shadow=True, startangle=140 )
plt.title("avocado percentage per type", bbox={'facecolor':'1', 'pad':8})

```

```python
#### average prices evolution per type from 2015 to 2018
avocado_avg_price = pd.pivot_table(avocado, index= "YEAR" , columns = "TYPE" , values = "AVERAGEPRICE" , aggfunc="mean")
avocado_avg_price
rcParams ["figure.figsize"] = 15,5
plt.plot(avocado_avg_price.conventional , label = 'conventional', linewidth=5.0)
plt.plot(avocado_avg_price.organic , label = 'organic', linewidth=5.0)
plt.grid(True)
plt.title("average prices evolution per type from 2015 to 2018" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("YEAR",color = "blue",fontweight = "bold")
plt.ylabel("PRICE", color = "blue", fontweight = "bold")
plt.legend()



```

```python
#### TOTAL VOLUME evolution per type from 2015 to 2018

avocado_volume = pd.pivot_table(avocado, index= "YEAR" , columns = "TYPE" , values = "TOTAL VOLUME" , aggfunc="sum")
print(avocado_volume)
rcParams ["figure.figsize"] = 15,5
plt.plot(avocado_volume.conventional , label = 'conventional', linewidth=5.0)
plt.plot(avocado_volume.organic , label = 'organic', linewidth=5.0)
plt.grid(True)
plt.title("TOTAL VOLUME evolution per type from 2015 to 2018" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("YEAR",color = "blue",fontweight = "bold")
plt.ylabel("VOLUME", color = "blue", fontweight = "bold")
plt.legend()

```

```python
##  create new column "Month"

avocado['MONTH'] = pd.to_datetime(avocado['DATE']).dt.month
avocado['MONTH']

```

```python
avocado["MONTH"].unique()
```

```python
### volume of avocado per month from 2015 to 2017


month_volume = pd.pivot_table(avocado, index= "MONTH" , columns = "YEAR" , values = "TOTAL VOLUME" , aggfunc="sum")
month_volume



```

**we should exclude year 2018 from graphic output becaue there is missing values to ensure the credibility of Analysis**

```python
##delete 2018 column
month_volume = month_volume.iloc[:, [0,1,2]]
```

```python
month_volume.plot(kind="bar")
plt.title("volume of avocado per month from 2015 to 2017" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("MONTH",color = "blue",fontweight = "bold")
plt.ylabel("VOLUME", color = "blue", fontweight = "bold")
plt.legend()
```

```python
### total volume evolution per hass from 2015 to 2019
volume_hass = avocado.groupby(["YEAR"]).agg( HASS_4046 = ('PLU_4046', 'sum') , HASS_4225 = ('PLU_4225', 'sum') , HASS_4770 = ('PLU_4770', 'sum' ))
rcParams ["figure.figsize"] = 15,5
plt.plot(volume_hass.HASS_4046 , label = 'HASS_4046', linewidth=5.0)
plt.plot(volume_hass.HASS_4225 , label = 'HASS_4225', linewidth=5.0)
plt.plot(volume_hass.HASS_4770 , label = 'HASS_4770', linewidth=5.0)
plt.grid(True)
plt.title("total volume evolution per hass from 2015 to 2019" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("YEAR",color = "blue",fontweight = "bold")
plt.ylabel("VOLUME", color = "blue", fontweight = "bold")
plt.legend()
```

```python
##total volume evolution per bag size from 2015 to 2019
volume_bag = avocado.groupby(["YEAR"]).agg( SMALL_BAGS = ('SMALL BAGS', 'sum') , LARGE_BAGS = ('LARGE BAGS', 'sum') , XLARGE_BAGS = ('XLARGE BAGS', 'sum' ))
from matplotlib import rcParams
rcParams ["figure.figsize"] = 15,5
plt.plot(volume_bag.SMALL_BAGS , label = 'SMALL BAGS',linewidth=5.0 )
plt.plot(volume_bag.LARGE_BAGS , label = 'LARGE BAGS', linewidth=5.0)
plt.plot(volume_bag.XLARGE_BAGS , label = 'XLARGE BAGS' ,linewidth=5.0)
plt.grid(True)
plt.title("total volume evolution per bag size from 2015 to 2019" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("YEAR",fontweight = "bold")
plt.ylabel("VOLUME",fontweight = "bold")
plt.legend()


```

```python
## top 10 most expensive region from 2015 to 2019 
top_expensive = avocado.groupby(["REGION"]).agg(AVERAGE_PRICE = ("AVERAGEPRICE" , "mean"))
top_expensive = top_expensive.sort_values(by='AVERAGE_PRICE', ascending=False)
top_expensive = top_expensive.head(10)
top_expensive = top_expensive.sort_values('AVERAGE_PRICE')
top_expensive.plot(kind="barh")
plt.title("top 10 most expensive region from 2015 to 2018" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("PRICE(DOLLARS)",fontweight = "bold")
plt.ylabel("REGION",fontweight = "bold")
plt.legend()
```

```python
## top 10 most cheap region from 2015 to 2019
top_cheap = avocado.groupby(["REGION"]).agg(AVERAGE_PRICE = ("AVERAGEPRICE" , "mean"))
top_cheap = top_cheap.sort_values(by='AVERAGE_PRICE', ascending=True)
top_cheap = top_cheap.head(10)
top_cheap = top_cheap.sort_values('AVERAGE_PRICE',ascending=False )
top_cheap.plot(kind="barh")
plt.title("top 10 most cheap region from 2015 to 2018" , color = "red" ,fontsize = 20, fontweight = "bold")
plt.xlabel("PRICE(DOLLARS)",fontweight = "bold")
plt.ylabel("REGION",fontweight = "bold")
plt.legend()
```
